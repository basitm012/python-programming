1) A company decided to give bonus of 5% to employee if his/her year of service is more than 5 years.
Ask user for their salary and year of service and print the net bonus amount.

```python
salary = int(input("Enter your salary amount: "))
appointed_year = int(input("Enter your Appointed Year: "))

if appointed_year<=2021:
    print("Your net bonus amount is Rs.",salary*5//100)
else:
    print("You're not eligibe for bonus amount")

print("Your salary is Rs.",salary)
print("You have been serving company from",appointed_year )
```

```python
num = int(input("Enter your number:"))

if num % 2 == 0:
    print("Your number is even")

else:
    print("Your number is odd")

print("Your number is", num)
```
