3. Find and display the largest number of a list without using built-in function max(). Your program
should ask the user to input values in list from keyboard.

```python
list = input("Enter your numbers seperated by comma:").split(',')

largest = list[0]

for num in list:
    if num > largest:
        largest = num

print ("The largest number among", list, "is", largest)
```

6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It
should print the date in the form March 12, 2021.

```python
date_input = input("Enter a date (mm/dd/yyyy): ")

month = date_input.split("/")[0]
date = date_input.split("/")[1]
year = date_input.split("/")[2]

months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

month_name = months[int(month) - 1]

print(f"{month_name} {date}, {year}")
```

8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be
like this:



```python
matrix = [[2, 11, 7, 12],
          [5, 2, 9,  15],
          [8, 3, 10, 42]]

for x in range(len(matrix)): 
    row_sum = 0
    for y in range(len(matrix[x])):
        row_sum += matrix [x][y]
    print(f"Sum of row {x} = {row_sum}")
```

9. Write a program to add two matrices of size n x m.

```python
lst_1 = [[10,20],
         [50,60],
         [100,200]]
lst_2 = [[30,40],
         [70,80],
         [300,400]]
result = []
answer = []

for x in range(len(lst_1)):
    for y in range(len(lst_1[x])):
        answer.append(lst_1[x][y] + lst_2[x][y])
    result.extend([answer])
    answer = []

print(result)
```
