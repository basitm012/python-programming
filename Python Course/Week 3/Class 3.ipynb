## Conditional Statement

- if internet available -> Class will be conducted
- else -> Class will not be conducted

`SYNTAX:`
`if <condition>:`
   `<body of if>`
   `else:`
   `<class will not be conducted>`

```python
marks = 25

if marks > 80:
    print("Your Grade is A")
elif marks > 60 and marks <= 80:
    print("Your Grade is B")
elif marks > 50 and marks <= 60:
    print("Your Grade is C")
elif marks > 45 and marks <= 50:
    print("Your Grade is D")
elif marks >= 25 and marks <= 45:
    print("Your Grade is E")
elif marks < 25:
    print("Your Grade is F")
```

```python
number = 4

if number % 2 == 0:
    print("even")
else:
    print("odd")
```

```python
number = int(input ("Enter Your Number: "))

if number % 2 == 0:
    print("Your number type is Even")
else:
    print("Your number type is *Odd*")

print ("Your number is", number) 
```

## Concatenation:
- For Concatenation using `+` operation, you `have` to type Cast
- For Concatenation using `f` operation, you `don't need` to type Cast

```python
fname = "Abdul"
lname = "Basit"
```

```python
fname + " " + lname
```

#### Method 2: Using F-String

```python
f"{fname} {lname}"
```

```python
fullname = f"{fname} {lname}"
```

```python
print(fullname)
```
