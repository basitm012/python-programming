### Tuples

`````python
[1,2,3,4,5,6,7,8,9,0] #list --> mutuale --> changeable
(1,2,3,4,5,6,7,8,9,0) #tuple --> immuteable --> not changeable
`````

`````python
type([1,2,3,4,5,6,7,8,9,0])
`````

`````python
type((1,2,3,4,5,6,7,8,9,0))
`````

`````python
list = [1,2,3,4,5,6,7,8,9,0]
list.append (11)
list[0] = 10 #update number
list
`````

`````python
# Because of immuutabily, tuple will give ERROR on updation  
tpl = (1,2,3,4,5,6,7,8,9,0) #do not have append function. it is unchanegable
## tpl.append (11) #'tuple' object has no attribute 'append'
## tpl [0] = 10 #'tuple' object does not support item assignment
`````

`````python
tpl[2:5] #same principle of slicing of list
`````

`````python
tpl[1::2]
`````

`````python
tpl[::-1]
`````

`````python
tpl = (1,2,3,4,5,6,7,8,9,0)

tpl[1::2] #even
tpl[::2]  #odd
tpl[::-1] #reverse
`````

## **Loops**

```bash
for <temp-variable> in <iterator>:
    <for loop body>
````

`````python
tpl
`````

`````python
tpl[0]
`````

`````python
for temp in tpl:
    print(temp)
`````

`````python
for temp in tpl[::-1]:
    print(temp)

#can use any variable in place of temp
`````
